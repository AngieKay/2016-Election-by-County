```python
df = pd.read_csv('ak_2012_2016.csv')
```

```python
missing_columns = pd.read_csv('alaska_missing_columns.csv')
```

```python
df.head()
```

```python
df = df.drop(df.iloc[:, [4, 6, 7, 8, 11]], axis = 1)
```

```python
df.head()
```

```python
df.columns = [col.replace(',', '') for col in df.columns]
```

```python
df = df.rename(columns = {'Registered Voters': '2016_registered_voters','Weighted/Muni':'County', 'Clinton Hillary':'Clinton', 'Trump Donald J.':'Trump', 'WtTotal':'2016_total_votes', 'Registered Voters .1':'2012_registered_voters', 'Total Votes':'2012_total_votes', 'Obama/Biden (DEM)':'Obama', 'Romney/Ryan (REP)':'Romney'})
```

```python
conditions = [df.iloc[:, 2] < df.iloc[:, 3], df.iloc[:, 2] > df.iloc[:, 3]]
choices = ['Trump', 'Clinton']
df['Target'] = np.select(conditions, choices, default = np.nan)
```

```python
df
```

```python
df.drop(df.index[29], inplace = True)
```

```python
df = df.replace({'Lake Peninsula':'Lake Peninsula', 'Wade Hampton': 'Kusilvak', 'Kodiak Island': 'Kodiak Isl', 'Lake and Peninsula': 'Lake Peninsula'}, regex = True)
```

```python
df = df.drop(df.iloc[:, 2:4], axis = 1)
```

```python
df['State'] = 'Alaska'
```

```python
df = df.sort_values(by = 'County')
```

```python
df.head()
```

```python
missing_columns.replace('Wade Hampton(kusilvak)', 'Kusilvak', inplace = True)
missing_columns = missing_columns.drop(missing_columns.iloc[:, 4:6], axis = 1)
```

```python
missing_columns = missing_columns.sort_values(by = 'Weighted/Muni')
```

```python
missing_columns
```

```python
missing_columns
```

```python
alaska_results = df.reset_index(drop=True).merge(missing_columns.reset_index(drop=True), left_index=True, right_index=True)
```

```python
alaska_results = alaska_results.drop(alaska_results.iloc[:, [1, 3, 9]], axis = 1)
```

```python
alaska_results.rename(columns = {'Land Area': '2010_land_area', 'Metro/Nonmetro status': 'central_outlying'}, inplace = True)
```

```python
alaska_results
```

```python
dump(alaska_results, open('alaska_final.pkl', 'wb'))
```
